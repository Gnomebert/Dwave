### $Distance^2$ for Ambulance at position marked 'A0' if it services every location in a 5*5 grid
M(5*5) = $\begin{bmatrix} 
2 & 1 & 2 & 5 & 10  \\ 
1 & A0 & 1 & 4 & 9  \\ 
2 & 1 & 2 & 5 & 10  \\ 
5 & 4 & 5 & 8 & 13  \\ 
10& 9 & 10 &13 & 18  \\ 
\end{bmatrix}$ 

Each element in M, row i {0,4} column j, is the $Distance_{ij}^2$ from the location of A0, where i=1,j=1.
So if A0 is the only Ambulance, the distance component of energy in the ising problem is given by;

Distance energy = $ \sum_{ij}^{all}{Distance_{ij}^2}$

The leading diagonal adds up to 30

Diagonal = $ \sum_{ij}^{i=4-j}{Distance_{ij}^2}=10+4+2+4+10=30$

Above the diagonal = $ \sum_{ij}^{i<4-j}{Distance_{ij}^2}=20$

Below the diagonal = $ \sum_{ij}^{i>4-j}{Distance_{ij}^2}=100$
## Optimal solution when A0 and A1 are on the same diagonal
For A0 at 'A0' and A1 at 'A1', the optimal allocation of destinations is shown below as '1' for A0, '0' for A1, and either for 'x'

A0_fix = 6 ie (1,1) A1_fix = 18  ie (3,3)

$\begin{bmatrix} 
1 & 1 & 1 & 1 & x  \\ 
1 & A0 & 1 & x & 0  \\ 
1 & 1 & x & 0 & 0  \\ 
1 & x & 0 & A1 & 0  \\ 
x & 0 & 0 & 0 & 0  \\ 
\end{bmatrix}$ 


This will have energy of 30 which ever is allocated the diagonal, another 20 for A0 allocated to all above the diagonal, and a final 20 for all those allocated to A1 below the diagonal. Energy of ground state = 30+20+20 = 70
